```python
# Importing the NumPy library for numerical operations
import numpy as np

# Importing the pandas library for data manipulation and analysis
import pandas as pd

# Importing matplotlib's pyplot module for data visualization
import matplotlib.pyplot as plt

# This magic command is used in Jupyter notebooks to display plots inline (inside the notebook)
%matplotlib inline

# Importing the seaborn library for statistical data visualization (built on top of matplotlib)
import seaborn as sns
```

```python
# Reading the CSV file named 'Diwali Sales Data.csv' into a DataFrame named 'df'
# The 'encoding="unicode_escape"' is used to handle special characters in the file (like ₹ or accented letters)
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.head() # it will print the top five values of datas
```

```python
# Returns a tuple representing the shape of the DataFrame: (number of rows, number of columns)
df.shape
```

```python
# Displays concise summary of the DataFrame including:
# - Total number of entries (rows)
# - Number of columns
# - Column names and data types
# - Non-null values count for each column (useful to detect missing data)
df.info()
```

```python
# Dropping the columns 'Status' and 'unnamed1' from the DataFrame
# axis=1 indicates columns (axis=0 would be for rows)
# inplace=True makes the change directly in the existing DataFrame/ it save the data 
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
# Checks for missing (null) values in each column of the DataFrame
# Returns the total number of null values for each column
pd.isnull(df).sum()
```

```python
# It will the Drop null Values
df.dropna(inplace = True)
```

```python
# Changing the data type of the 'Amount' column to integer (int)
# This is useful for performing numerical operations like sum, average, etc.
df['Amount'] = df['Amount'].astype('int')
```

```python
#check the data type
df['Amount'].dtypes
```

```python
#print the all columns
df.columns
```

```python
# Rename the column
# Its not save beacause of not use inplace = True
df.rename(columns = {'Marital_Status':'Shaadi'})
```

```python
# Generates descriptive statistics of numerical columns in the DataFrame
# Includes count, mean, standard deviation, min, max, and quartile values (25%, 50%, 75%)
df.describe()
```

## Exploratory Data Analysis

```python
df.columns
```

#### Gender

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Grouping the data by 'Gender' and calculating the total 'Amount' spent by each gender
# as_index=False ensures 'Gender' remains a column instead of becoming the index
# sort_values is used to sort the result in descending order of total 'Amount'
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Creating a bar plot using Seaborn to visualize total sales amount by gender
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
# Barplot - It used for the comparing two values or cateogory 
```

###### From above graphs we can see that most of the buyers are females and even the purchasing power of female is greater than the male

```python
# Countpot - Used for the count of columns
```

```python
# Create a count plot using Seaborn to show the number of customers in each 'Age Group',
# separated by 'Gender' using different colors (hue = 'Gender')
ax = sns.countplot(data=df, x='Age Group', hue='Gender')

# Loop through each set of bars (Male & Female for each Age Group) and add value labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_gen = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_gen)
```

##### From above graphs we can see that most of the buyers are of age group between 26-35 years females.

#### State

```python
df.columns
```

```python
# Grouping data by 'State' and calculating total number of orders per state
# as_index=False keeps 'State' as a column (not an index)
# Sorting states by total 'Orders' in descending order
# .head(10) gets only the top 10 states with the most orders
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Setting figure size for the plot (15 inches wide, 5 inches tall)
sns.set(rc={'figure.figsize': (17, 5)})

# Creating a barplot of total orders for the top 10 states
sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
# Grouping the data by 'State' and calculating the total sales amount for each state
# as_index=False ensures 'State' remains a column (not index)
# Sorting the result in descending order of 'Amount' (highest sales first)
# .head(10) picks the top 10 states by sales
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Setting the size of the plot (15 inches wide, 5 inches tall)
sns.set(rc={'figure.figsize': (17, 5)})

# Creating a barplot to visualize total sales amount from the top 10 states
sns.barplot(data=sales_state, x='State', y='Amount')
```

##### From the avove graphs we can see that most of the orders and Total sales/amount are from Uttar Pradesh, Maharastra and Karnataka respectively

#### Marital Status

```python
# Set the figure size for the plot (6 inches wide, 5 inches tall)
sns.set(rc={'figure.figsize': (6, 5)})

# Create a count plot of the 'Marital_Status' column (e.g., Married vs Unmarried)
# This shows how many customers fall into each marital status category
ax = sns.countplot(data=df, x='Marital_Status')

# Add count labels (number of people) on top of each bar
for bars in ax.containers:
    ax.bar_label(bars)
```

##### From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

#### Occupation

```python
ax = sns.countplot(data = df,x = 'Occupation')

sns.set(rc={'figure.figsize': (25, 5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount')

sns.set(rc = {'figure.figsize':(25,7)})
sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount')
```

##### From above graphs we can see that most of the buyers are working in It,Healthcare and Aviation

#### Produt Category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False) ['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y = 'Amount')
```

##### From above graphs we can see that sold products are from clothing and Electronics category

```python
sales_state = df.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc ={'figure.figsize':(20,5)})

sns.barplot(data = sales_state, x = 'Product_ID',y = 'Orders')
```
